```python
import requests
from bs4 import BeautifulSoup
import json
import time
from datetime import datetime
import os


# CONFIG

NEWS_URL = "https://www.moneycontrol.com/news/business/"
JSON_FILE = "moneycontrol_news.json"
REFRESH_TIME = 10  # seconds

HEADERS = {
    "User-Agent": "Mozilla/5.0"
}
```

```python
# LOAD EXISTING NEWS

def load_existing_news():
    if os.path.exists(JSON_FILE):
        with open(JSON_FILE, "r", encoding="utf-8") as f:
            return json.load(f)
    return []
```

```python
# SAVE NEWS TO JSON

def save_news(news_list):
    with open(JSON_FILE, "w", encoding="utf-8") as f:
        json.dump(news_list, f, indent=4, ensure_ascii=False)
```

```python
# SCRAPE MONEYCONTROL

def scrape_moneycontrol():
    response = requests.get(NEWS_URL, headers=HEADERS, timeout=10)
    soup = BeautifulSoup(response.text, "html.parser")

    articles = soup.find_all("li", class_="clearfix")
    scraped_news = []

    for article in articles:
        title_tag = article.find("h2")
        link_tag = article.find("a")

        if not title_tag or not link_tag:
            continue

        headline = title_tag.text.strip()
        link = link_tag.get("href")

        timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

        scraped_news.append({
            "headline": headline,
            "link": link,
            "timestamp": timestamp
        })

    return scraped_news
```

```python
# REMOVE DUPLICATES

def remove_duplicates(existing_news, new_news):
    existing_links = {item["link"] for item in existing_news}

    unique_news = []
    for item in new_news:
        if item["link"] not in existing_links:
            unique_news.append(item)

    return unique_news
```

```python
# MAIN LOOP

def main():
    print("🔄 Moneycontrol News Scraper Started...\n")

    while True:
        try:
            existing_news = load_existing_news()
            new_scraped_news = scrape_moneycontrol()
            unique_news = remove_duplicates(existing_news, new_scraped_news)

            if unique_news:
                updated_news = existing_news + unique_news
                save_news(updated_news)
                print(f"✅ {len(unique_news)} new news articles saved at {datetime.now()}")
            else:
                print(f"⏳ No new news at {datetime.now()}")

            time.sleep(REFRESH_TIME)

        except Exception as e:
            print("❌ Error:", e)
            time.sleep(REFRESH_TIME)
```

```python
# RUN

if __name__ == "__main__":
    main()
```
